# # Class Distribution Analysis

```python
plt.figure(figsize=(8, 6))
sns.countplot(x='Class', data=creditcard_df)
plt.title('Credit Card Fraud Distribution')
plt.show()
print(creditcard_df['Class'].value_counts(normalize=True))
```

# Feature Correlation

```python
plt.figure(figsize=(15, 10))
corr = creditcard_df.corr()
sns.heatmap(corr, cmap='coolwarm', annot=False)
plt.title('Correlation Heatmap')
plt.show()
```

# Transaction Amount Analysis

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x='Class', y='Amount', data=creditcard_df[creditcard_df['Amount'] < 3000])
plt.title('Transaction Amount by Class (filtered < 3000)')
plt.show()
```
